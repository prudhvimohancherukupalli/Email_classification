```
traindata=datasets.fetch_20newsgroups(subset='train')
testdata=datasets.fetch_20newsgroups(subset='test')
x_train=traindata.data  #extacting all the mails from dataset
y_train=traindata.target #extrcting lables
names=traindata.target_names #extrcting names
print(len(names))
```

```
for i in range(len(x_train)):
  doc =x_train[i]
  doc=re.sub('[a-zA-Z0-9._]+@[a-zA-z0-9._]+','',doc)
  doc=re.sub('[0-9]+[0-9a-zA-Z._]+','',doc)
  doc=re.sub('_','',doc)
  
  x_train[i]=doc
```

```
import spacy
nlp=spacy.load('en_core_web_sm')
def tokenizer(text):
  doc=nlp(text)
  doc=[token.lemma_.lower().strip() for token in doc]
  doc=[token for token in doc if token!='-pron-']
  doc=[token for token in doc if token.isalpha()]
  return doc
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(lowercase=True,stop_words='english',tokenizer=tokenizer)
tfidf.fit(x_train)
print(tfidf.get_feature_names())
print(len(tfidf.get_feature_names()))
```

```
x_train=tfidf.transform(x_train).toarray()
x_train.shape
```

```
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
from sklearn import metrics
metrics.accuracy_score(testdata.target,model.predict(tfidf.transform(testdata.data).toarray()))
```
